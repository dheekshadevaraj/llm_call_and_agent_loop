```python
from openai import OpenAI

client = OpenAI()

response = client.responses.create(
    model="gpt-4o-mini",
    input="Hello Agent! What can you do?"
)

print(response.output_text)
```

```python
from openai import OpenAI

client = OpenAI()

tasks = ["search weather", "do math", "summarize text"]

for task in tasks:
    print(f"\n[Agent received task]: {task}")

    response = client.responses.create(
        model="gpt-4o-mini",
        input=[
            {"role": "system", "content": "You are a helpful autonomous agent."},
            {"role": "user", "content": f"Please {task}"}
        ]
    )

    print("[Agent output]:", response.output_text)
```

```python
response = client.responses.create(
    model="gpt-4o-mini",
    input=[
        {"role": "system", "content": "You are an agent that explains your reasoning step by step."},
        {"role": "user", "content": "Solve 27 * 43"}
    ]
)

print(response.output_text)
```
